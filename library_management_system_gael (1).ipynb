```python
# Python Library Management System

# Install required libraries (This should already exist in colab)
!pip install requests

import pickle
import sqlite3
import requests

# ===============================
# 1. Data Types and Variables
# ===============================

book_title = "Python Programming"
book_quantity = 5
book_price = 29.99
is_available = True

# Display book information
print(f"Book: {book_title}, Quantity: {book_quantity}, Price: ${book_price}, Available: {is_available}")


# ===============================
# 2. Lists
# ===============================

books = ["Python Basics", "Advanced Python", "Data Science with Python"]

# Add two more books to the list
books.append("Machine Learning Basics")  # Adds at end
books.insert(1, "Deep Learning Fundamentals")  # Inserts at index 1

print("Updated book list:", books)


# ===============================
# 3. Loops and range()
# ===============================

print("\nAvailable books:")

# Loop through books with indices
for i in range(len(books)):
    print(f"{i}: {books[i]}")


# ===============================
# 4. User Input
# ===============================

new_book = input("Enter a new book: ")

# Validate input before adding
if new_book.strip() != "":
    books.append(new_book.strip())
    print("New book added:", books)
else:
    print("Invalid input. No book added.")


# ===============================
# 5. Functions and Conditional Statements
# ===============================

def check_availability(book_name):
    """
    Check if a book exists in the list.
    Returns True if available, False otherwise.
    """
    if book_name in books:
        return True
    else:
        return False


# Test the function
print(f"Is 'Python Basics' available? {check_availability('Python Basics')}")


# ===============================
# 6. File I/O
# ===============================

def save_books_to_file(filename):
    """
    Save the list of books to a text file.
    """
    try:
        with open(filename, 'w') as f:
            for book in books:
                f.write(f"{book}\n")
        print("Books saved successfully.")
    except Exception as e:
        print(f"Error writing file: {e}")


def read_books_from_file(filename):
    """
    Read books from a file and return them as a list.
    Handles file errors.
    """
    loaded_books = []
    try:
        with open(filename, 'r') as f:
            for line in f:
                book = line.strip()
                if book:
                    loaded_books.append(book)
    except FileNotFoundError:
        print("File not found.")
    except Exception as e:
        print(f"Error reading file: {e}")

    return loaded_books


# Test file I/O
save_books_to_file("books.txt")
print("Books from file:", read_books_from_file("books.txt"))
!cat books.txt  # Display file contents


# ===============================
# 7. Pickle for serialization
# ===============================

def save_books_pickle(filename):
    """
    Save books list using pickle.
    """
    try:
        with open(filename, 'wb') as f:
            pickle.dump(books, f)
        print("Books saved as pickle.")
    except Exception as e:
        print(f"Error saving pickle: {e}")


def load_books_pickle(filename):
    """
    Load books list from a pickle file.
    """
    try:
        with open(filename, 'rb') as f:
            return pickle.load(f)
    except FileNotFoundError:
        print("Pickle file not found.")
    except Exception as e:
        print(f"Error loading pickle: {e}")

    return []


# Test pickle functions
save_books_pickle("books.pkl")
print("Books loaded from pickle:", load_books_pickle("books.pkl"))


# ===============================
# 8. SQLite Database Interaction
# ===============================

def create_books_table():

    """Create books table in SQLite database."""

    try:
        conn = sqlite3.connect('library.db')
        cursor = conn.cursor()

        cursor.execute('''CREATE TABLE IF NOT EXISTS books
                          (id INTEGER PRIMARY KEY,
                           title TEXT,
                           quantity INTEGER)''')

        conn.commit()
        conn.close()
        print("Database table created.")
    except sqlite3.Error as e:
        print(f"Database error: {e}")


def add_book_to_db(title, quantity):
    """
    Add a book to the SQLite database.
    """
    try:
        conn = sqlite3.connect('library.db')
        cursor = conn.cursor()

        cursor.execute(
            "INSERT INTO books (title, quantity) VALUES (?, ?)",
            (title, quantity)
        )

        conn.commit()
        conn.close()

        print(f"Book '{title}' added to database.")
    except sqlite3.Error as e:
        print(f"Database error: {e}")


# Test database functions
create_books_table()
add_book_to_db("Python for Beginners", 10)


# ===============================
# 9. Web API Interaction (GitHub API)
# ===============================

def get_python_repos():
    """
    Fetch top 5 Python repositories from GitHub.
    """
    url = "https://api.github.com/search/repositories"
    params = {"q": "language:python", "sort": "stars", "order": "desc"}

    try:
        response = requests.get(url, params=params)
        data = response.json()
        return [repo['name'] for repo in data['items'][:5]]
    except Exception as e:
        print(f"API Error: {e}")
        return []


# Test API function
print("Top 5 Python repositories on GitHub:", get_python_repos())


# ===============================
# Final Output
# ===============================

print("\n✅ Library Management System operations completed.")
```
